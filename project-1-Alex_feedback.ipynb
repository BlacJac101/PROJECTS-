```python
def compress_function(string):
    word={}
    result=" "
    for x in string:
        if x in word:
            word[x] = word[x]+1
        else:
            word[x] = 1
    for key, value in word.items():
        result += str(key) + str(value)

    return result


if __name__ == '__main__':
    s = input("Enter the string:")
    print(compress_function(s))

#My code is able to be utilised irrespective of if it be upper or lower case 
```
